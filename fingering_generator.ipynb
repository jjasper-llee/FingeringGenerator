```python
def cello(notes):
    # initialize the dynamic programming table
    dp = [[None] * 4 for _ in range(len(notes))] # 4 is the number of fingers on a hand
    dp[0][0] = {'fingers': [1], 'cost': 0}
    dp[0][1] = {'fingers': [2], 'cost': 0}
    dp[0][2] = {'fingers': [3], 'cost': 0}
    dp[0][3] = {'fingers': [4], 'cost': 0}


    # fill in the dynamic programming table
    best_fingers = []
    for i in range(1, len(notes)):
        for j in range(4):
            best_fingers = None
            best_cost = float('inf')

            for k in range(4):
              while notes[i] >= 12:
                notes[i] -= 12

              pos = find_pos(notes[i-1], k+1)
              if pos != 0:
                cur_pos = pos
              else:
                break

              cur_string = find_string(notes[i-1])
              num_half_steps = notes[i] - notes[i-1]

              # print("cur_pos: " + str(cur_pos))
              # print("num_half_steps " + str(num_half_steps))
              # print("cur finger: " + str(k+1) + "new finger: " + str(j+1))
              # print("cur_note: " + str(notes[i]))

              try:
                  if num_half_steps > 0:
                    c = cost_dict_shift_up[cur_pos][num_half_steps][(k+1, j+1)] # the cost of playing notes[i] with finger j having played notes[i-1] with finger k th
                  else:
                    c = cost_dict_shift_down[cur_pos][abs(num_half_steps)][(k+1, j+1)]
              except KeyError:
                print("KeyError")
                # break
                c = 10

              # print("c = {}".format(c))
              # print("dp cost = {}".format(dp[i-1][k]['cost']))
              total_cost = c + dp[i-1][k]['cost']
              # print("Total cost: {}".format(total_cost))
              # print("Best cost: {}".format(best_cost))
              if total_cost < best_cost:
                best_cost = total_cost # update the total cost of using the fingerings to reach this note thus far
                best_fingers = dp[i-1][k]['fingers'] + [j+1] # update the fingerings chosen to reach this note thus far
              dp[i][j] = {'fingers': best_fingers, 'cost': best_cost}

    # find the optimal fingering
    best_fingers = dp[-1][0]['fingers']
    best_cost = dp[-1][0]['cost']
    for i in range(1, 4):
        if dp[len(notes)-1][i]['cost'] < best_cost:
            best_fingers = dp[-1][i]['fingers']
            best_cost = dp[-1][i]['cost']
    print(best_fingers)
    return best_fingers

def find_pos(cur_note, cur_finger):
    pos_note = cur_note - (cur_finger - 1) #is the note the first finger is on
    if pos_note <= 0: return 0
    if pos_note % 12 == 1:
        return 0.5
    elif pos_note % 12 == 2:
        return 1
    elif pos_note % 12 == 3:
        return 2
    elif pos_note % 12 == 4:
        return 2.5
    elif pos_note % 12 == 5:
        return 3
    elif pos_note % 12 == 6:
        return 3.5
    elif pos_note % 12 >= 7:
        return 4

def find_string(cur_note):
  if cur_note <= 11: return 4
  elif cur_note <= 23: return 3
  elif cur_note <= 35: return 2
  elif cur_note <= 47: return 1
```

```python
def cello2(notes):

  best_fingers = []
  best_cost = float('inf')

  for start_finger in range(1,5):
    cur_best_fingers = [start_finger]
    cur_best_cost = 0
    previous_finger = start_finger

    # These two for loops should generate the optimal fingerings based on playing the first note with finger i i.e. they will run 4 times
    for note in range(1,len(notes)):
      if notes[note] % 12 == 0:
        cur_best_fingers.append(0)
        note += 1
      else:

        # finding the best finger given the curent note thus far
        best_new_finger_cost = float('inf')
        best_new_finger = 0

        # convert 1st note to one string system
        while notes[0] >= 12:
            notes[0] -= 12

        for cur_finger in range(1,5):
          cur_new_finger_cost = 0

          # convert all notes to one string system
          while notes[note] >= 12:
            notes[note] -= 12

          # looking for the position of the note
          pos = find_pos(notes[note-1], previous_finger)
          if pos != 0:
            cur_pos = pos
          else:
            cur_finger += 1

          # calculating the half steps between current note and the previous ntoe
          num_half_steps = notes[note] - notes[note-1]

          # finding the cost here to shift up or down
          try:
              if num_half_steps > 0:
                cur_new_finger_cost = cost_dict_shift_up[cur_pos][num_half_steps][(previous_finger, cur_finger)] # the cost of playing notes[i] with finger j having played notes[i-1] with finger k th
              else:
                cur_new_finger_cost = cost_dict_shift_down[cur_pos][abs(num_half_steps)][(previous_finger, cur_finger)]
          except KeyError:
            # print("KeyError")
            # break
            cur_new_finger_cost = 10

          print("cur_new_finger_cost: {}".format(cur_new_finger_cost))
          print("Cur finger: {}".format(cur_finger))
          print("cur note: {}".format(notes[note]))


          # finding the best finger to put down by comparing their individual costs
          if cur_new_finger_cost < best_new_finger_cost:
            best_new_finger_cost = cur_new_finger_cost
            best_new_finger = cur_finger

        # adding the best new finger cost and best new finger to the cur best fingers and cur_best_cost starting on finger i
        cur_best_cost += best_new_finger_cost
        cur_best_fingers.append(best_new_finger)
        previous_finger = best_new_finger
        # print("Cur pos: {}".format(cur_pos))
        # print("Cur best fingers: {}".format(cur_best_fingers))
        # print("Cur best cost: {}".format(cur_best_cost))
        #break

    # updating global best cost and fingerings  by comparing cur best cost
    if cur_best_cost < best_cost:
      best_cost = cur_best_cost
      best_fingers = cur_best_fingers
      print("Best Cost: {}".format(cur_best_cost))
      print("Best Fingers: {}".format(cur_best_fingers))

  return best_fingers

def find_pos(cur_note, cur_finger):
    pos_note = cur_note - (cur_finger - 1) #is the note the first finger is on
    if pos_note <= 0: return 0
    if pos_note % 12 == 1:
        return 0.5
    elif pos_note % 12 == 2:
        return 1
    elif pos_note % 12 == 3:
        return 2
    elif pos_note % 12 == 4:
        return 2.5
    elif pos_note % 12 == 5:
        return 3
    elif pos_note % 12 == 6:
        return 3.5
    elif pos_note % 12 >= 7:
        return 4
```
